```python
from googleapiclient.discovery import build
import pandas as pd

API_KEY = 'AIzaSyCHHgJgdSuECAIOfs0fuRcD8nbhVMnfamE'  # <-- Replace with your API key
YOUTUBE_API_SERVICE_NAME = 'youtube'
YOUTUBE_API_VERSION = 'v3'

youtube = build(YOUTUBE_API_SERVICE_NAME, YOUTUBE_API_VERSION, developerKey=API_KEY)

# Search for tech channels (using 'technology' as a keyword)
search_response = youtube.search().list(
    q='python programming',
    type='channel',
    part='id',
    maxResults=50
).execute()

channel_ids = [item['id']['channelId'] for item in search_response['items']]

# Get channel statistics
channels_response = youtube.channels().list(
    id=','.join(channel_ids),
    part='snippet,statistics',
    maxResults=50
).execute()

channels = []
for item in channels_response['items']:
    channels.append({
        'title': item['snippet']['title'],
        'subscribers': int(item['statistics'].get('subscriberCount', 0)),
        'channelId': item['id']
    })

# Sort by subscribers
channels_sorted = sorted(channels, key=lambda x: x['subscribers'], reverse=True)
top_10 = channels_sorted[:10]

df = pd.DataFrame(top_10)[['title', 'subscribers', 'channelId']]
df
```
